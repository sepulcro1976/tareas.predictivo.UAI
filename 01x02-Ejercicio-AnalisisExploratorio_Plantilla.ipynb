## Descripción dataset
Un conjunto de datos completo que captura la naturaleza multifacética de los ciberataques en 225 países durante un período de 14 meses desde octubre de 2022 hasta diciembre de 2023. El conjunto de datos, que comprende 77.623 filas y 18 campos, proporciona una cronología detallada de los ciberataques, categorizados en ocho críticos. dimensiones. A continuación se muestran las descripciones de las columnas:

* Attack-Date (Fecha): Registra la fecha de cada ciberataque. Las 77.623 entradas son válidas, lo que indica un conjunto de datos completo sin fechas faltantes.
* País (Texto): Especifica el país donde se produjo el ciberataque. Todas las entradas son válidas, cubren 225 países y brindan una perspectiva global sobre las amenazas cibernéticas.
* Spam (Decimal): registra el porcentaje de ataques relacionados con spam para un país específico en un día particular (valor porcentual en decimal). Los datos de muestra que se muestran en la Tabla 3 del paper (ver en aula) para Spam son 0,01358, lo que significa que Bangladesh sufrió el 1,358% de los ataques de Spam ese día. Con 62.982 entradas válidas, la media es 0,006094 y la desviación estándar es 0,024337. El rango va desde un mínimo de 0,000010 hasta un máximo de 0,302490. Los percentiles 25, 50 y 75 son 0,000090, 0,000590 y 0,003530, respectivamente.
* Ransomware (Decimal): registra el porcentaje de ataques relacionados con ransomware para un país específico en un día determinado. Consta de 52.144 entradas válidas. El valor medio es 0,000130, con una desviación estándar de 0,000186. El rango abarca desde 0,000010 hasta 0,009180. Los percentiles son 0,000040 (25), 0,000070 (50) y 0,000140 (75).
* Infección local (decimal): registra el porcentaje de infecciones locales para un país específico en un día particular. Con 74.469 entradas, la media es 0,013350 y la desviación estándar es 0,008415. Los valores oscilan entre 0,000240 y 0,049370. Los percentiles son 0,007150 (25), 0,010790 (50) y 0,017660 (75).  
* Exploit (Decimal): registra el porcentaje de ataques relacionados con Exploit para un país específico en un día particular. Contiene 64.264 entradas válidas. La media es 0,000469, con una desviación estándar de 0,000368. El rango es de 0,000010 a 0,004660. Los percentiles 25, 50 y 75 son 0,000210, 0,000390 y 0,000620.
* Correo malicioso (decimal): registra el porcentaje de ataques relacionados con correo malicioso para un país específico en un día particular. Comprende 69.184 entradas. La media es 0,001292, con una desviación estándar de 0,001606. El rango abarca desde 0,000010 hasta 0,043220. Los percentiles son 0,000300 (25), 0,000730 (50) y 0,001690 (75).
* Ataque a la red (decimal): registra el porcentaje de incidentes relacionados con ataques a la red para un país específico en un día determinado. Con 71.532 entradas, el valor medio es 0,002222 y la desviación estándar es 0,003034. El rango es de 0,000020 a 0,058260. Los percentiles 25, 50 y 75 son 0,000700, 0,001290 y 0,002350.
* Análisis bajo demanda (decimal): registra el porcentaje de análisis bajo demanda (realizados por el programa antivirus) para un país específico en un día determinado. Contiene 74.231 entradas válidas. La media es 0,009756, con una desviación estándar de 0,006080. Los valores mínimo y máximo son 0,000240 y 0,124880, respectivamente. Los percentiles son 0,005210 (25), 0,008140 (50) y 0,012870 (75).
* Amenaza web (decimal): registra el porcentaje de ataques relacionados con amenazas web para un país específico en un día determinado. Comprende 73.892 entradas.El valor medio es 0,013006, con una desviación estándar de 0,004943. El rango abarca desde 0,000240 hasta 0,048630. Los percentiles 25, 50 y 75 son 0,009700, 0,012570 y 0,015973.  
* 11-18 -- Campos de clasificación (texto): estos campos (clasificación de spam, clasificación de ransomware, clasificación de infección local, clasificación de explotación, clasificación de correo malicioso, clasificación de ataque de red, clasificación de escaneo bajo demanda, clasificación de amenaza web) proporcionan la clasificación mundial de cada país en el respectivo dimensión del ciberataque. Las clasificaciones varían de 1 a un máximo de 196, y las clasificaciones medias varían ligeramente según los diferentes tipos de ataques. La desviación estándar en estas clasificaciones indica una dispersión moderada, lo que refleja el impacto variable de las amenazas cibernéticas en los diferentes países.
